```python
import matplotlib.pyplot as plt

# Generate sample data for x
x_values = range(35, 80)  # for x from 0 to 20

# Compute y values using y = 43000 + 1000 * x
y_values = [43000 + 1000*x for x in x_values]

# Create the plot
plt.plot(x_values, y_values)  # Plot the regression line
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Model: y = 43000 + 1000x')
plt.show()
```
